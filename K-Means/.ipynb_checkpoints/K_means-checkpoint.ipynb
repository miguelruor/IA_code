# Algoritmo de clusterización: K-means

```python
import pandas as pd 
import numpy as np
import math
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
np.random.seed(0)
```

### Extracción de datos: dataset en https://archive.ics.uci.edu/ml/datasets/Sales_Transactions_Dataset_Weekly#

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00396/Sales_Transactions_Dataset_Weekly.csv"
X = pd.read_csv(url, header = 0)
#y = X.iloc[:,-1] # la variable independiente se encuentra en la ultima columna
X.drop(X.columns[0], axis='columns', inplace=True)
n = len(X.columns)
X.drop([X.columns[i] for i in range(52, n)], axis = 'columns', inplace = True)
#X.columns = range(X.shape[1])
print(X)
```

### Distancia Minkowski de $x,y$ vectores en $\mathbb{R}^n$

```python
def MinkowskiDistance(x,y, p=2): # Distancia Minkowski de x,y vectores en R^n
  dist = 0
  n = len(x)

  for i in range(0,n):
    dist += abs(x[i]-y[i])**p

  dist = dist**(1/p)

  return dist
```

### Implementación del K-means

```python
class kMeans:
  def __init__(self, k, maxIter, tol):
    self.k = k
    self.maxIter = maxIter
    self.tol = tol

  def fit(self, X):
    # maxIter es el numero maximo de iteraciones
    # tol es la tolerancia de convergencia
    self.X = X.copy(deep = True)
    k = self.k
    maxIter = self.maxIter
    tol = self.tol

    n = X.shape[0]
    dim = X.shape[1]

    ind = np.random.choice(n, k, replace=False) # muestra aleatoria de tamaño k sin reemplazo de los numeros 0, 1, ..., n-1. 
    # Estos indices determinarian los centroides iniciales

    labels = [0 for i in range(n)]

    centroids = [np.array(X.iloc[ind[i]]) for i in range(k)] # arreglo de los centroides

    for t in range(maxIter): 
      for i  in range(n):
        distMin = math.inf # infinito
        indMin = 0

        for j in range(k):
          d = MinkowskiDistance(centroids[j], np.array(X.iloc[i]))

          if(d < distMin): # si encontramos una nueva distancia menor
            indMin = j # actualizamos la etiqueta del patron i como etiqueta j
            distMin = d
        
        labels[i] = indMin # asignamos al patron i al cluster indMin con la distancia minima

      centroidsAux = [np.zeros(dim) for i in range(k)]

      for j in range(k):
          patternsj = X.iloc[[i for i in range(n) if labels[i] == j]] # patrones en el cluster j 
          centroidsAux[j] = np.array(patternsj.sum()/len(patternsj))
        
      # Necesitamos ver si el algoritmo ya convergio
      flag = True # Esta variable nos ayudara a saber si el algoritmo ya convergio o no

      for j in range(k):
        d = MinkowskiDistance(centroids[j], centroidsAux[j])
        if d > tol:
          flag = False # Aun no converge el algoritmo
          break

      centroids = centroidsAux

      if flag == True: #Los centroides ya no cambian mas de la tolerancia tol, i.e., tenemos convergencia del algoritmo
        self.labels = pd.Series(labels)
        self.centroids = centroids
        return
    
      self.labels = pd.Series(labels)
      self.centroids = centroids

  def predict(self, X_new):
    n = X_new.shape[0]
    labels = []

    for i  in range(n):
      distMin = math.inf # infinito
      indMin = 0

      for j in range(k):
        d = MinkowskiDistance(self.centroids[j], np.array(X_new.iloc[i]))

        if(d < distMin): # si encontramos una nueva distancia menor
          indMin = j # actualizamos la etiqueta del patron i como etiqueta j
          distMin = d
        
      labels.append(indMin) # asignamos al patron i al cluster indMin con la distancia minima
    
    return labels

  def silhouette(self):
    clusters = [] # lista para guardar los clusters con sus patrones respectivos
    n = self.X.shape[0]

    for j in range(self.k):
      patternsj = self.X.iloc[[i for i in range(n) if self.labels[i] == j]] # patrones en el cluster j
      clusters.append(patternsj)

    self.silhouette_coef = [] # lista para guardar los coeficientes de silueta

    for i in range(n):
      ai = 0
      labeli = self.labels[i] # cluster al que pertenece el patron i
      di = clusters[labeli].shape[0]

      if di == 1:
        self.silhouette_coef.append(0)
        continue

      for j in range(di): # calcular la distancia promedio del patron i a todos los patrones en el cluster labeli
        ai += MinkowskiDistance(np.array(self.X.iloc[i]), clusters[labeli].iloc[j])/(di-1)
      
      bi = math.inf

      for j in range(self.k): # Encontrar la distancia promedio del patron i a todos los patrones en el cluster mas cercano al labeli
        if j == labeli:
          continue

        dj = clusters[j].shape[0] 
        bi_aux = 0

        for l in range(dj):
          bi_aux  += MinkowskiDistance(np.array(self.X.iloc[i]), clusters[j].iloc[l])/dj

        if bi_aux < bi:
          bi = bi_aux
      
      
      si = (bi-ai)/max(ai, bi)
      self.silhouette_coef.append(si)

    return
```

### Coeficiente de Silueta

Supongamos que el patrón $i$ está en el cluster $C_i$. Para calcular el coeficiente de silueta de $i$, tenemos que calcular $a(i)$ la distancia promedio de $i$ a todos los otros patrones en $C_i$, y $b(i)$ la distancia promedio de $i$ a todos los patrones en el cluster más cercano a $C_i$, es decir,
$$
a(i) = \frac{1}{|C_i|-1}\sum_{j\in C_i} d(i,j), \quad b(i) = \min_{k\neq i} \frac{1}{|C_k|}\sum_{j\in C_k} d(i,j).
$$
De esta manera, el coeficiente de silueta del patrón sería 
$$
s(i) = \begin{cases}
0, & |C_i| = 1, \\
\frac{b(i)-a(i)}{max(a(i),b(i))}, & |C_i| > 1.
\end{cases}
$$

### Aplicar K-Means a nuestros datos

```python
k = 5
maxIter = 100
tol = 10**(-2)
```

```python
clustering = kMeans(k, maxIter, tol)
clustering.fit(X)
print(clustering.labels)
print(clustering.centroids)
```

```python
clustering.silhouette()
clustering.silhouette_coef
```

### Verificación con un ejemplo

```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets.samples_generator import make_blobs 
%matplotlib inline
```

Generamos 4 clusters artificialmente para ver si el algoritmo los encuentra

```python
X, y = make_blobs(n_samples=500, centers=[[2,2], [-5, -2], [-8, 8], [10,-5]], cluster_std=0.9)
```

```python
plt.scatter(X[:, 0], X[:, 1], marker='.')
```

Aplicamos K-means

```python
X = pd.DataFrame(X)
clustering = kMeans(4, maxIter, tol)
clustering.fit(X)
X = X.to_numpy()
labels = clustering.labels
centroids = clustering.centroids
```

Graficamos los patrones coloreados de acuerdo al cluster al que pertenecen. Para generar la gráfica se utilizó el código de este repositorio: https://github.com/anandprabhakar0507/python-K-means-clustering/blob/master/ML0101EN-Clus-K-Means-Customer-Seg-py-v1.ipynb

```python
fig = plt.figure(figsize=(6, 4))
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(labels)))) # 4 colores
ax = fig.add_subplot(1, 1, 1)

for k, col in zip(range(4), colors):

    my_members = (labels == k)
    centroid = centroids[k]
    
    ax.plot(X[my_members, 0], X[my_members, 1], 'w', markerfacecolor=col, marker='.')
    ax.plot(centroid[0], centroid[1], 'o', markerfacecolor=col,  markeredgecolor='k', markersize=6)

plt.show()
```

Podemos ver que el algoritmo sí funciona como debería. Esto se puede comprobar con los coeficientes de silueta que son cercanos a 1.

```python
clustering.silhouette()
clustering.silhouette_coef
```
